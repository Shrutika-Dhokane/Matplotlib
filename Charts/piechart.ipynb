```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["home rent","food","phone bill","car","other utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%',shadow=True,explode=[0,0.1,0.1,0,0],startangle=45)
plt.show()
plt.savefig("piechart.png",bbox_inches='tight',pad_inches=2,transparent=True)
```
